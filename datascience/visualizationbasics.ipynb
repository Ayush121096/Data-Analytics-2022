```python
x2 = x ** 2
```

```python
plt.plot(x,x2)
```

```python
x3 = x**3
```

```python
plt.figure(figsize=(15,7))
plt.plot(x, label='original' ,linestyle='--')
plt.plot(x2, label='square' ,linestyle=':')
plt.plot(x3, label='cubed')
plt.legend()# activate the point(legend)
plt.xlabel("data points")# add a label too the x axis
plt.ylabel("values of x")# add a label to the y axis
plt.title('the visualization of x axis and its square and cube')
plt.show()

```

```python
cities = ['lucknow','delhi','mumbai']
rainfall = [39 ,20, 10]
```

```python

plt.bar(cities, rainfall)
```

```python
plt.bar(cities, rainfall, color = 'yellow' , edgecolor='black' , hatch='//' , align='center') # h stands for horizontal
plt.title('cities rainfall data')# hatch stands for pattern
```

```python
plt.figure(facecolor= 'white')
wedges = [1,3,4]
plt.pie(rainfall, labels=cities, autopct='%.1f%%' ,radius=2)
plt.show()
```
