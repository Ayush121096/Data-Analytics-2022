```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
# to show whole columns otherwise show some columns
```

```python
url = 'https://raw.githubusercontent.com/digipodium/Datasets/main/classfication/pokemon_type.csv'
df = pd.read_csv(url)
df.head()
```

```python
df['Name'] # select single column
```

```python
df.Type_1 # another way to acess a column
```

# way to acess a multiple columns


```python
cols = ['Total', 'HP', 'Attack']
df[cols]
```

```python
df [['Name','Total']] # df = data frame
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x='Name', y='HP', data=df[:50])
plt.xticks(rotation=90)# here you can add title, name as well
plt.show()
# xticks use to fix direction
```

```python

plt.figure(figsize=(20,5))
sns.barplot(x='Name',y='HP',data=df.sort_values('HP')[-100:])
plt.xticks(rotation=90)
plt.show()
# sort use for sorting the value from starting or from last
```

```python
# to check how 2 numerical columns are related
plt.figure(figsize=(10,10))
sns.scatterplot(x='Attack',y='Defense',data=df, hue='Type_2')
```

```python
sns.lmplot(x='Attack' ,y='Defense' ,data=df)
# implot is same to check how 2 numeric column is related
```

```python

# use to compare categorical(non numeric) column is related to a numerical column
sns.violinplot(x='Type_2' ,y='Defense' ,data=df)
plt.figure(figsize=(25,10))
```

```python
plt.figure(figsize=(25,6))
sns.swarmplot(x='Type_1',y='Attack',data=df,alpha=0.8)
# alpha use for transeparency
```

```python
sns.pairplot(df.select_dtypes(include=['int']),kind='reg')
```

```python
# visualize only categorical columns
sns.countplot(x='Type_1',data=df)# count use to show only particular column
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x='Body_Style',data=df)# put the value which you want to on x,y axis
plt.xticks(rotation=90)
plt.show()
```

```python
sns.jointplot(x='Attack',y='Defense',data=df,palette='cool',hue='Type_1')
# joint plot is use to show to graphs in one diagram
```

```python
sns.jointplot(x='Attack',y='Defense',data=df,kind='reg')
```

filtering a dataframe


```python
df['Type_1'] == 'Grass' # boolean filter
# search type_1 grass value,,True means yes and False means No
```

```python
df[ df['Type_1'] == 'Grass' ]
```

```python
grass_df = df[ df['Type_1'] == 'Grass' ].copy()
```

```python
sns.jointplot(x='Attack',y='Defense',data=grass_df, hue='Type_2',)
```

```python
f1  = df['Type_1'] == 'Grass' # boolean filter
f2  = df['Type_1'] == 'Fire' # boolean filter
df[f1 | f2]
```

```python
import plotly.express as px
```

```python
px.scatter_3d(df[f1 | f2], x='Attack', y='Defense',
              z='Speed', color='Type_2', size='Total',
              title='Pokemon Type Fire and Grass',
              height=600,width=600,hover_name='Name')
```
